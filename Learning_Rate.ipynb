# Learning Rate Demonstration
## $ g(x)=x^4-4x^2+5 $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make some data the x's to be thrown into g(x) and dg(x), we call x_2
x_2 = np.linspace(-2, 2, 1000)

# the g(x) function and the dg(x) function in Python
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
#Gradient decent function definition
def gradient_descent( initial_guess,  multiplier, precision, max_iter):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [dg(new_x)]

    for n in range(max_iter):
        previous_x = new_x 
        new_x = previous_x - multiplier * dg(previous_x)
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(dg(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
n = 100
local_min, list_x, deriv_list = gradient_descent(initial_guess=1.9, multiplier = 0.25, precision =0.0001, max_iter = n) 
# Plot function and derivative and scatter plot side by side

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.grid()
plt.xlim(-2,2)
plt.ylim(-0.5, 6.5)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.5)
plt.scatter(list_x, g(np.array(list_x)), color='green', s=100, alpha=0.9)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
   
print('Number of steps:', len(list_x))
print('Local min occurs at:', local_min)
```

## multiplier or learning rate or gamma = 0.0001

```python
n = 100
local_min, list_x_1, deriv_list = gradient_descent(initial_guess=3, multiplier = 0.0001, precision =0.0001, max_iter = n)
plt.figure(figsize=[20, 10])
plt.xlim(0, n)
plt.ylim(0, 50)
plt.title('Effect of the learning rate 1', fontsize=17)
plt.xlabel('Nr of iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
iteration_list = list(range(0, n+1))
plt.plot(iteration_list, g(np.array(list_x_1)), color='pink', linewidth=5)
plt.scatter(iteration_list, g(np.array(list_x_1)), color='red', s=80)
print('ITERATION LIST:',  len (iteration_list),     '.......LIST OF X\'s:', len(list_x_1),'.... MIN IS AT', list_x_1[100])
plt.show()
```
