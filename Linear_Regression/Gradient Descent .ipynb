# A Simple Cost Function

# $f(x)= x^2+x+1 $

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1=np.linspace(start=-3,stop=3,num=300)
#x_1
```

```python
def f(x):
    return x**2 + x + 1
```

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize =(8,5))
plt.plot(x_1,f(x_1))
plt.title('Simple cost function',fontsize = 18)
plt.xlabel( 'X', fontsize = 18)
plt.ylabel ('f(x)',fontsize = 18 )
plt.xlim(-3,3)
plt.ylim(0,8)
plt.show()
```

```python
plt.figure(figsize =(8,5))
plt.plot(x_1, df(x_1))
plt.title('Derivative of the Simple cost function',fontsize = 18)
plt.xlabel( 'X', fontsize = 18)
plt.ylabel ('df(x)',fontsize = 18 )
# plt.xlim()
# plt.ylim()
plt.show()
```

```python
# Gradient Descent.
new_x = 3
previous_x = 0
gamma = 0.1

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(100):
    previous_x = new_x
    new_x = previous_x - gamma*df(previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
```

```python
#print(x_list)
```

```python
#print(slope_list)
```

```python
plt.figure(figsize=[8, 5])
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.9)
values = np.array(x_list)
plt.scatter(x_list, f(values), color = 'red', s =100, alpha = 0.4)
plt.show()
```

```python
print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))  
```

## Example 2 - Multiple Minima vs Initial Guess & Advanced Functions

# $$ g(x)=x^44x^2+5 $$

```python
x_2 = np.linspace(-2, 2, 1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def dg(x):
    return 4*x**3 -8*x
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3)

plt.subplot(1, 2, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6,8)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5)
plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.01, precision=0.001,  max_iter=100):       
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break 
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=1.9)

plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)
#plt.xlim(-2, 2)
#plt.ylim(0.5, 5.5)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, g(np.array(list_x)), color='red', s=100, alpha=0.6)

plt.subplot(1, 2, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
#plt.xlim(-2, 2)
#plt.ylim(-6, 8)
plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)
plt.show()
```

```python
print('Number of steps:', len(list_x))
print('Local min occurs at:', local_min)
```

# $$ Example.3- h(x)=x^5-2x^4+2 $$

```python
import numpy as np
import matplotlib.pyplot as plt
x_3=np.linspace(start=-2.5,stop=2.5,num=1000)
```

```python
def h(x):
    return x**5 - 2*x**4 + 2
```

```python
def dh(x):
    return 5*x**4 - 8*x**3
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.02, precision=0.001,  max_iter=50):    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    for n in range(max_iter):  
        previous_x = new_x 
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        if step_size < precision:
            break 
    return new_x, x_list, slope_list
```

```python
# Calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess= 0.2, 
                                                max_iter=500)
# Plot function and derivative and scatter plot side by side

plt.figure(figsize=[15, 5])
# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.title('Cost function', fontsize=17)
#plt.xlabel('X', fontsize=16)
#plt.ylabel('h(x)', fontsize=16)
plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.6)
    
# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.title('Slope of the cost function', fontsize=17)
#plt.xlabel('X', fontsize=16)
#plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
```

```python
print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(list_x))
```

```python
#Calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess= -0.1, 
                                                max_iter=500)
# Plot function and derivative and scatter plot side by side

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)
plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)
plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.plot(x_3, dh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.show()

print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(list_x))
```
