```python
#Import Statements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Cost Revenue Problem

```python
data=pd.read_csv("cost_revenue_clean.csv")
data.head()
```

```python
data.describe()
```

```python
1.100000e+03
```

```python
X=pd.DataFrame(data,columns=["production_budget_usd"]) #Captial X because its a feature (Convection)
y=pd.DataFrame(data,columns=["worldwide_gross_usd"])
```

```python
plt.scatter(X,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3) 
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression=LinearRegression()
regression.fit(X,y)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.2, c = 'blue')
plt.plot(X, regression.predict(X), color='brown', linewidth=2) #For the Straight line Plt X against reg.predict(X)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget ')
plt.ylabel('Worldwide Gross ')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#Slope or theta_1
regression.coef_[0][0]
```

```python
#Intercept or theta_0
regression.intercept_[0]
```

```python
#Regression Score
regression.score(X,y)
```

```python
#y=mx+c
y=regression.coef_[0][0]*3.5e8 + regression.intercept_[0]
y
```

## Drug PPM Problem

```python
data1=pd.read_csv("lsd_math_score_data.csv")
data1.head()
```

```python
time = data1 [['Time_delay']]
LSD = data1 [['Drug_ppm']]
score = data1 [['Math_score']]
```

```python
regression1=LinearRegression()
regression1.fit(LSD,score)
```

```python
regression1.coef_[0][0]
```

```python
regression1.intercept_[0]
```

slope of the regression line = -9.009 and the intercept of the regression line is 89.123

```python
predictedscore=regression1.predict(LSD)
```

```python
predictedscore  #Adjusted Points after Linear regression
```

```python
score
```

```python
#Plot Of LSD and the Score
plt.figure(figsize=(15,6))
plt.scatter(LSD, predictedscore)
plt.plot(LSD, predictedscore)
plt.scatter(LSD, score)
plt.plot(LSD, score)
plt.show()
```

## Heart Disease Prediction

```python
data2=pd.read_csv("heart_disease_dataset.csv")
data2.head()
```

```python
data2.describe()
```

```python
X2=pd.DataFrame(data2,columns=["Age"]) #Captial X because its a feature (Convection)
y2=pd.DataFrame(data2,columns=["Cholesterol"])
```

```python
plt.scatter(X2,y2)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X2, y2, alpha=0.3) 
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.ylim(130, 600)
plt.xlim(30, 80)
plt.show()
```

```python
regression2=LinearRegression()
regression2.fit(X2,y2)
```

```python
regression2.intercept_[0]
```

```python
regression2.coef_[0][0]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X2, y2, alpha=0.2, c = 'blue')
plt.plot(X2, regression2.predict(X2), color='brown', linewidth=2) #For the Straight line Plt X against reg.predict(X)
plt.title('Age vs Cholesterol')
plt.xlabel('Age ')
plt.ylabel('Cholesterol')
plt.ylim(130, 600)
plt.xlim(30, 80)
plt.show()
```

```python
#Regression Score
regression2.score(X2,y2)
```

```python
y=regression2.coef_[0][0] * 67 + regression2.intercept_[0]
y
```
